```python
# We're going to convert a circuit to a gate
from qiskit import QuantumCircuit, QuantumRegister
q = QuantumRegister(5)
qc = QuantumCircuit(q)
```

```python
# let's apply some random gates
qc.h(q[0])
qc.h(q[4])
qc.mct([q[0], q[1], q[3]], q[4])  # multi-controlled gate, control: q0, q1, q3, target: q4
qc.swap(q[2], q[3])
qc.cy(q[0], q[4])
qc.draw('mpl')
```

```python
# let's create an inverse circuit where the gates are applied in inverse order
inverse_circuit = qc.inverse()
```

```python
inverse_circuit.draw('mpl')
```

```python
# let's construct a quantum gate with the inverse_circuit
my_gate = inverse_circuit.to_gate()
```

```python
q = QuantumRegister(7)
qc1 = QuantumCircuit(q)
```

```python
# let's now apply the custom gate above to the new circuit
qc1.append(my_gate, [q[1], q[2], q[3], q[4], q[5]])
```

```python
qc1.draw('mpl')
```
