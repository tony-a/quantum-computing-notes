```python
# Toffoli Gate (CCX)
from qiskit import QuantumCircuit, QuantumRegister
q = QuantumRegister(5)
qc = QuantumCircuit(q)
```

```python
from qiskit.quantum_info import Statevector
```

```python
psi1 = Statevector(qc)
psi1.draw('latex')
```

```python
# let's apply a Toffoli gate with q0 & q1 as control
# and target on q2
# this gate should do nothing now, because both 
# q0 & q1 are in the ground state |0⟩
qc.ccx(q[0], q[1], q[2])
qc.draw('mpl')
```

```python
psi2 = Statevector(qc)
psi2.draw('latex')
```

```python
# if we apply NOT gate on q0 & q1
# then apply the Toffoli gate
# we'll see q2 flip as well
qc.x(q[0])
qc.x(q[1])
qc.draw('mpl')
```

```python
psi3 = Statevector(qc)
psi3.draw('latex')
```

```python
qc.ccx(q[0], q[1], q[2])  # now apply Toffoli
qc.draw('mpl')
```

```python
psi4 = Statevector(qc)
psi4.draw('latex')
```

```python
# Let's implement a multi-controlled Toffoli gate
qc.mct(control_qubits=[q[0], q[2], q[3]], target_qubit=q[4])
qc.draw('mpl')
```

```python
# this multi-controlled gate won't flip q4
# since q3 is still |0⟩
psi5 = Statevector(qc)
psi5.draw('latex')
```

```python
# Let's apply multi-controlled Rotation_Y gate
from math import pi
```

```python
qc.mcry(pi/2, [q[1], q[2], q[3], q[4]], q[0])
qc.draw('mpl')
```
