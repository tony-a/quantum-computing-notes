```python
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
q = QuantumRegister(5, name='qubit')
c = ClassicalRegister(5, name='bits')
qc = QuantumCircuit(q, c)
```

```python
qc.draw('mpl')
```

```python
# let's apply some random quantum gates
# just to demonstrate different visualizations
for i in range(5) :
    qc.x(q[i])
    qc.h(q[i])

qc.barrier()
for i in range(2):
    qc.y(q[i])
    
qc.barrier()
for i in range(2, 5):
    qc.z(q[i])
```

```python
qc.draw('mpl')
```

```python
# we can also draw the circuit in 'text' form
qc.draw('text')
```

```python
# or  in 'latex_source'
print(qc.draw('latex_source'))
```

```python
qc.draw('mpl', reverse_bits=True)  # reverses the order of the qubits
```

```python
# we can also change the color of the gates
qc.draw('mpl', style='iqx')
```

```python
# to draw without plotting the barriers
qc.draw('mpl', plot_barriers=False)
```

```python
# changing scale
qc.draw('mpl', scale=0.5)
```

```python
# we can show the initial states
qc.draw('mpl', initial_state=True)
```

```python
qc.draw('mpl', cregbundle=False)  # this flag will seperate the classical bits to individual lines
```

```python
qc.measure_all()
```

```python
qc.draw('mpl', cregbundle=False)
```
