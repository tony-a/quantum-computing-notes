```python
# Initiate four quantum circuits with two qubit register each
# to build each of the four Bell's states
from qiskit import QuantumRegister, QuantumCircuit
q = QuantumRegister(2)
qc0 = QuantumCircuit(q)
qc1 = QuantumCircuit(q)
qc2 = QuantumCircuit(q)
qc3 = QuantumCircuit(q)
```

```python
from qiskit.quantum_info import Statevector
```

```python
# Let's construct Bell state Phi+
qc0.h(q[0])
qc0.cx(q[0], q[1])
qc0.draw('mpl')
```

```python
phi_plus = Statevector(qc0)
phi_plus.draw('latex')
```

```python
# Let's construct Bell state Phi-

qc1.h(q[0])
qc1.cx(q[0], q[1])
qc1.z(q[0])
qc1.draw('mpl')
```

```python
phi_minus = Statevector(qc1)
phi_minus.draw('latex')
```

```python
# Let's construct Bell state Psi+

qc2.h(q[0])
qc2.cx(q[0], q[1])
qc2.x(q[1])
qc2.draw('mpl')
```

```python
psi_plus = Statevector(qc2)
psi_plus.draw('latex')
```

```python
# Let's construct Bell state Psi-
qc3.h(q[0])
qc3.cx(q[0], q[1])
qc3.z(q[0])
qc3.x(q[1])
qc3.draw('mpl')
```

```python
psi_minus = Statevector(qc3)
psi_minus.draw('latex')
```

```python
# Draw all 4 Bell state circuits together
display(qc0.draw('mpl'), qc1.draw('mpl'), qc2.draw('mpl'), qc3.draw('mpl'))
```

```python
# Let's visualize the Bloch sphere to demonstrate 
# that these two qubits are entangled and therefore
# can not be decomposed into individual qubits in 
# a Bloch sphere. 
# The visualization should return blank spheres
from qiskit.visualization import plot_bloch_multivector
```

```python
plot_bloch_multivector(qc0)
```

```python
plot_bloch_multivector(qc1)
```

```python
print(phi_plus.purity())  # a purity of 1 means the state is NOT mixed.
```

```python
# Let's examine the density matrix
from qiskit.quantum_info import DensityMatrix
```

```python
rho_plus = DensityMatrix(qc0)
rho_plus.draw('latex', prefix='\\rho_{AB}=')
```

```python
# Let's find the "reduced density matrix" now
from qiskit.quantum_info import partial_trace
```

```python
rho_A = partial_trace(rho_plus, [0])
rho_B = partial_trace(rho_plus, [1])
```

```python
rho_A.draw('latex', prefix='\\rho_A=')
```

```python
rho_B.draw('latex', prefix='\\rho_B=')
```

```python
# calculate the purity of rho_A and rho_B
rho_A.purity()  # a purity of 1/2 means the states are completely mixed
```

```python
rho_B.purity()  # a purity of 1/2 means the states are completely mixed
```
